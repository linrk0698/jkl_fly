Use pyplot interface to create figure, then use object methods for the rest

```python
def my_plotter(ax,data1,data2,param_dict): 
    out = ax.plot(data1,data2,**param_dict)
    return out

data1,data2,data3,data4 = np.random.randn(4,100)
fig,ax = plt.subplots(1,1)
my_plotter(ax,data1,data2,{'marker':'x'})
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
my_plotter(ax1,data1,data2,{'marker':'x'})
my_plotter(ax2,data3,data4,{'marker':'o'})
```

```python
mpl.rc('lines',linewidth='4',color='g')
plt.plot(data)
```

## Working with multiple figures and axes 

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

## Working with text 
The `text()` command can be used to add text in an arbitrary location. 

```python
mu,sigma = 100,15
x = mu + sigma * np.random.randn(10000)

n,bins,patches = plt.hist(x,50,density=1,facecolor='g',alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,0.025, r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

## Annotating text 

```python
ax = plt.subplot(111)

t = np.arange(0.0,5.0,0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t,s,lw=2)
plt.annotate('local max',xy=(2,1),xytext=(3,1.5),arrowprops=dict(facecolor='black',shrink=1))
plt.ylim(-2,2)
plt.show()
```
