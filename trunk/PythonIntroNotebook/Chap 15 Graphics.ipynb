```python
# 15.3.2 Multiple plots on the same Axz
x = np.random.randn(100)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(x,bins = 30, label = 'Empirical')
xlim = ax.get_xlim()
ylim = ax.get_ylim()
pdfx = np.linspace(xlim[0],xlim[1],200)
pdfy = stats.norm.pdf(pdfx)
pdfy = pdfy/pdfy.max() * ylim[1]
plt.hold(True)
plt.plot(pdfx,pdfy,'r-',label = 'PDF')
ax.set_ylim((ylim[0],1.2*ylim[1]))
plt.legend()
plt.hold(False)
#plt.draw()
```

```python
#15.3.4 Dates on Plots
import datetime as dt

T = 2000
x = []
for i in range(T):
    x.append(dt.datetime(2012,3,1)+dt.timedelta(days = i))
y = np.cumsum(np.random.randn(T))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
fig.autofmt_xdate()
```

```python
import matplotlib.dates as mdates 
T = 100
x = []
for i in range(1,T+1):
    x.append(dt.datetime(2012,3,1)+dt.timedelta(i,0,0))
y = np.cumsum(np.random.randn(T))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)

months = mdates.MonthLocator()
ax.xaxis.set_major_locator(months)
fmt = mdates.DateFormatter('%b %Y')
ax.xaxis.set_major_formatter(fmt)
fig.autofmt_xdate()

xlim = list(ax.get_xlim())
xlim[0] = mdates.date2num(dt.datetime(2012,3,1))
ax.set_xlim(xlim)

```

```python
# TEX in Plots
from matplotlib import rc 
rc('text',usetex = True)
rc('font',family = 'serif')
y = 50 * np.exp(.0004 + np.cumsum(.01 *np.random.randn(100)))
plt.plot(y)
plt.xlabel(r'time($\tau$)')
plt.ylabel(r'Price',fontsize = 16)
plt.title(r'Geometric Random Walk: $d\ln p_t = \mu dt + \sigma dW_t$',fontsize = 16)
rc('text',usetex = False)
```

```python
# 15.4.1 Line Plots

from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0,6*np.pi,600)
z = x.copy()
y= np.sin(x)
x = np.cos(x)
fig = plt.figure()
ax = Axes3D(fig) # Essential Line. 
ax.plot(x,y,zs = z, label = 'Spiral')
ax.view_init(15,45)
```

```python
#15.4.2 Surface and Mesh Plots
import numpy.linalg as linalg

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
x,y = np.meshgrid(x,y)
```

```python
z = np.mat(np.zeros(2))
p = np.zeros(np.shape(x))
R = np.matrix([[1,0.5],[0.5, 1]])
Rinv = linalg.inv(R)
for i in range(len(x)):
    for j in range(len(y)):
        z[0,0] = x[i,j]
        z[0,1] = y[i,j]
        p[i,j] = 1.0/(2*np.pi) * np.sqrt(linalg.det(R))*np.exp(-(z*Rinv*z.T)/2)
```

```python
np.shape(p)
```

```python
p[1,1]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(x,y,p,rstride=5,cstride=5,color='#AD5300')
ax.view_init(29,20)
```

```python
import matplotlib.cm as cm
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x,y,p,rstride = 2,cstride = 2,cmap = cm.coolwarm,shade = 'interp')
ax.view_init(29,80)
plt.savefig('figure.pdf')
plt.savefig('figure.png')
```
